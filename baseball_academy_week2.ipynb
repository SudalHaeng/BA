>사용자 정의 함수(def) 만들기


함수 만들기 : 결과물을 재사용
def로 모듈을 생성, 모듈을 하나의 폴더에 구성해놓은 것

ex) import pandas (panday.py라는 모듈을 사용하라)


불러오기 : import ______

사용     : 모듈이름.함수이름      모듈이름.변수이름

```python
#def 함수 만들기 : 결과물을 재사용
def df(Ryu, Toronto):
    if Ryu==True:
        print("류현진")
    else:
        print("류현진이 아님")
        if Toronto==True:
            ("토론토")
        else:
            print("토론토 아님")
```

```python
Ryu=True
Toronto=True

df(Ryu,Toronto)
```

```python
Ryu=False
Toronto=True

df(Ryu,Toronto)
```

```python
Ryu=False
Toronto=False

df(Ryu,Toronto)
```

```python
#삼진 72개 중 구종별 구속 최대값
df2.groupby('pitch_type').release_speed.max()
```

```python
df2.groupby('pitch_type')['release_speed'].max()
```

```python
#삼진 72개 중 구종별 구속 평균값
df2.groupby('pitch_type')['release_speed'].mean()
```

```python
#삼진 72개 중 구종별 개수
df2.groupby('pitch_type').count()
```

```python
df2.groupby('pitch_type').count()['game_date']
```

```python
#구종별 스피드 
df2.groupby('pitch_type')['release_speed'].agg(['mean','min','max','count'])
```

```python
#구종별 스피드, 회전수
df2.groupby('pitch_type')['release_speed','release_spin_rate'].agg(['mean','min','max','count'])
```

```python

#김광현의 메이저리그에서 삼진의 구종별 구속 평균, 최고와 개수를 구하라

```

```python
#피벗테이블로도 가능
df.pivot_table(values='release_speed', index='pitch_type', aggfunc=['max', 'min', 'mean'])
df.pivot_table(values=['release_speed','release_spin_rate'], index='pitch_type', aggfunc=['max', 'min', 'mean'])
```


* values : 계산에 쓰일 칼럼 (궁금한 것)
* index  : 왼쪽줄(행), 그룹으로 묶을 기준이 되는 칼럼
* column : 윗줄(열)에 쓰이는 분류
* aggfunc: 계산 대상 값에 적용할 내용 / mean,max,min 등 

```python

#새로운 컬럼 생성하기 , 공격 팀
df2.loc[df2['inning_topbot']=='Top','Offense_Team']=df2.loc[df2['inning_topbot']=='Top','away_team']
df2.loc[df2['inning_topbot']=='Bot','Offense_Team']=df2.loc[df2['inning_topbot']=='Bot','home_team']


```
