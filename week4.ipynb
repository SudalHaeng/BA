```python
def strike_zone(player, alpha):
  #류현진
  df2 = df[(df['pitcher']==player)]

  # 스윙과 콜스트라이크 구분
  S = df2.loc[(df2.description == 'swinging_strike') | (df2.description == 'swinging_strike_blocked')]  #헛스윙
  B = df2.loc[df2.description == 'called_strike']  #스트라이크 콜

  #로케이션 박스 
  L, R = -0.708333, +0.708333
  Bot, Top = 1.5, 3.5

  plt.figure(figsize=(4,4), dpi=100)

  x2 = [L, R, R, L, L]
  y2 = [Top, Top, Bot, Bot, Top]

  #홈플레이트 (포수시점)
  L_p = L+0.1
  L_m = L-0.1
  R_p = R+0.1
  R_m = R-0.1
  Center=0

  S_height = 0 
  M_height = -0.6
  E_height = -1.0

  x = [R_m, L_p, L_m, Center, R_p, R_m]
  y = [S_height, S_height, M_height, E_height, M_height, S_height]

  #x,y 최대 범위
  plt.xlim([L-2.5, R+2.5])
  plt.ylim([Bot-3, Top+2])
  plt.title('투구 위치 그리기')


  #그래프 그리기 (scatter plot+박스+홈플레이트)
  plt.scatter(S['plate_x'], S['plate_z'], s=50, color='red',  alpha=alpha, label='헛스윙')
  plt.scatter(B['plate_x'], B['plate_z'], s=50, color='blue', alpha=alpha, label='콜스트라이크')
  plt.plot(x2, y2, c='black', lw=2.5)
  plt.plot(x, y, c='grey', lw=0.5)

  plt.show()
  plt.close()
```

```python
# 직구를 500구 이상 던진 00명의 선수들의 헛스윙 비율
# 규칙에 의한 스트라이크 존 안에 들어왔을 경우에만

df3=df[(df['pitch_type']=='FF') & (df['plate_x']> -0.708333) & (df['plate_x']< 0.708333) &
        (df['plate_z']> 1.5) & (df['plate_z']< 3.5)]

df3=df3.assign(swinging_strike=  # 헛스윙 했을 때                                                                                                 
         np.where(df3['description'].isin(['swinging_strike','swinging_strike_blocked']),1,0))

df3=df3.assign(total=  # 전체 투구
         np.where(df3['description'].isin(df3['description'].unique()),1,0))

df3=df3.groupby('pitcher')['total','swinging_strike'].sum()
df3['swinging_per']=df3['swinging_strike']*100/df3['total']
df3=df3[df3['total']>200]
df3
```

```python
#트레킹 데이터 평균 요약 831명 
df3_=df.groupby('pitcher')['pfx_z','release_speed','effective_speed','release_spin_rate','release_extension','plate_z'].mean()
df3_['BU']=df3_['release_spin_rate']/df3_['release_speed']
df3_


#두 데이터를 레프트 조인으로 결합
result=pd.merge(df3,df3_,on='pitcher',how='left')
result=result[['pfx_z','plate_z', 'release_speed', 'effective_speed',
       'release_spin_rate', 'release_extension', 'BU','swinging_per']]
result
```

```python
#상관관계 행렬
result.corr(method='pearson')


#산점도 행렬
sns.pairplot(result, diag_kind='hist')
plt.show()
```

```python
#상관 매트릭스

def corrdot(*args, **kwargs):
    corr_r = args[0].corr(args[1], 'pearson')
    corr_text = f"{corr_r:2.2f}".replace("0.", ".")
    ax = plt.gca()
    ax.set_axis_off()
    marker_size = abs(corr_r) * 10000
    ax.scatter([.5], [.5], marker_size, [corr_r], alpha=0.6, cmap="coolwarm",
               vmin=-1, vmax=1, transform=ax.transAxes)
    font_size = abs(corr_r) * 40 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction",
                ha='center', va='center', fontsize=font_size)

sns.set(style='white', font_scale=1.6)


g = sns.PairGrid(result, aspect=1.4, diag_sharey=False)
g.map_lower(sns.regplot, lowess=True, ci=False, line_kws={'color': 'black'})
g.map_diag(sns.distplot, kde_kws={'color': 'black'})
g.map_upper(corrdot)
```

```python
df = pd.read_csv("/content/drive/MyDrive/baseball_academy/mlb_pbp_2021.csv")
```

```python
Zone_In = df[((df.plate_x > -0.708333) & (df.plate_x < 0.708333) & 
           (df.plate_z > 1.5) & (df.plate_z < 3.5 ) ) &
           (df.description.isin(["called_strike","ball","blocked_ball"]))]

Zone_Out=  df[~((df.plate_x > -0.708333) & (df.plate_x < 0.708333) & 
           (df.plate_z > 1.5) & (df.plate_z < 3.5 ) ) &
           (df.description.isin(["called_strike","ball","blocked_ball"]))]
```

```python
#가상의 스트라이크 존 안에서 심판이 콜스트라이크 할 확률 
len(Zone_In[(Zone_In.description=="called_strike")])*100/len(Zone_In)
```

```python
#가상의 스트라이크 존 밖에서 심판이 볼콜 할 확률 
len(Zone_Out[Zone_Out.description.isin(["ball","blocked_ball"])])*100/len(Zone_Out)
```
